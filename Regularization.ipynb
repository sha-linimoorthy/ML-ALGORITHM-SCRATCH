**Ridge Regression**

L2 Regularization, adds penality equivalent to the square of the magnitude of coefficients Minimization objective, penalty term=> squared Ridge
image.png

**Lasso Regression**

Performs L1 Regularization


image.png

Ridge=> add squared penalty to loss

Lasso=> add absolute penalty to loss




**Ridge** can not reach to 0, in local minima, closer to Zero (shape circle)

**Lasso** can reach to zero as we use ABS (shape diamond)

Overfit overcome Regularize

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.array([1,2,3,4,5])
y=np.array([2.1,3.8,6.5,9.2,11.3])
```

```python
X_b=np.c_[np.ones((X.shape[0],1)),X]
```

```python
X_b
```

```python
alpha_ridge=1.0
alpha_lasso=0.1
num_iteration=1000
learning_rate=0.01
```

```python
def lasso_regression(X,y,alpha,num_iterations,learning_rate):
  m,n=X.shape
  theta=np.zeros(n)
  for _ in range(num_iterations):
    y_pred=X.dot(theta)
    gradient=(X.T.dot(y_pred-y)+alpha * np.sign(theta)) /m #np.sign -1 x<0 0x==0 1 x>0
    theta-=learning_rate*gradient
  return theta
```

```python
def ridge_regression(X,y,alpha,num_iterations,learning_rate):
  m,n=X.shape
  theta=np.zeros(n)
  for _ in range(num_iterations):
    y_pred=X.dot(theta)
    gradient=(X.T.dot(y_pred-y)+alpha * theta) /m
    theta-=learning_rate*gradient
  return theta
```

```python
lasso=lasso_regression(X_b,y,alpha_lasso,num_iteration,learning_rate)
```

```python
ridge=ridge_regression(X_b,y,alpha_ridge,num_iteration,learning_rate)
```

```python
print(lasso)
```

```python
print(ridge)
```

```python
plt.scatter(X,y,label="Data")
plt.plot(X,X_b.dot(lasso),label=f'Lasso L1 Regression',color='red')
plt.plot(X,X_b.dot(ridge),label=f'Ridge L2 Regression',color='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
