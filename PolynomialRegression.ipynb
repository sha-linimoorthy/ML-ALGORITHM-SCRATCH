```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.array([1,2,3,4,5])
Y=np.array([1,8,27,64,125])
```

```python
degree=2
```

```python
X_poly=np.column_stack([X**i for i in range(degree+1)])
```

```python
print(X_poly)
```

**Add Weights**

```python
weights=np.zeros(degree+1)
weights
```

**Gradient Descend**

 **Iteration**

```python
learning_rate=0.001
num_iteration=10000
```

```python
for _ in range(num_iteration):
  prediction=np.dot(X_poly,weights)
  Error=prediction-Y
  gradient=np.dot(X_poly.T,Error)/len(X) #change in Weights with respect to error
  weights-=learning_rate * gradient
  print(f'Weights for iteration {_},{weights}')
```

```python
def predict(x,weights):
  x_poly=np.array([x**i for i in range(degree+1)])
  return np.dot(x_poly,weights)
```

```python
plt.scatter(X,Y,label="Data Points")
x_range=np.linspace(min(X),max(X),100)
y_pred=[predict(x,weights) for x in x_range]
print(y_pred)
plt.plot(x_range,y_pred,color='r',label="polynomial Regression")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
